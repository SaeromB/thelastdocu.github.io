```python
recommend = pd.read_csv("data/recommendation_result.csv", encoding = 'utf-8')
recommend.head()
```

```python
jsondata = []
    
for i in range(len(recommend)):
   
    for_json= []

    for_json.append('{')
    for_json.append('"index" : "{}",'.format(recommend.loc[i]['index']))
    
    reco_list = []
    reco_list.append("{}".format(recommend.loc[i]['reco1']))
    reco_list.append("{}".format(recommend.loc[i]['reco2']))
    reco_list.append("{}".format(recommend.loc[i]['reco3']))
    reco_list.append("{}".format(recommend.loc[i]['reco4']))
    reco_list.append("{}".format(recommend.loc[i]['reco5']))
    reco_list.append("{}".format(recommend.loc[i]['reco6']))
    reco_list.append("{}".format(recommend.loc[i]['reco7']))
    reco_list.append("{}".format(recommend.loc[i]['reco8']))
    reco_list.append("{}".format(recommend.loc[i]['reco9']))
    reco_list.append("{}".format(recommend.loc[i]['reco10']))
    
    for_json.append('"reco" : "{}",'.format(reco_list))
    for_json.append('}')
    
    for_json = ''.join(for_json)
    
    jsondata.append(for_json)

jsondata = ','.join(jsondata)

```
